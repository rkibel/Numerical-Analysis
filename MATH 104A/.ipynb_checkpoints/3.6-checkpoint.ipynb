```python
# 3.6

import math
import numpy as np
from sympy import *
from sympy.plotting import plot
from sympy.plotting.plot import MatplotlibBackend, Plot
import matplotlib
%matplotlib inline 
# %matplotlib ipympl
# %matplotlib tk 

import matplotlib.pyplot as plt

t = symbols('t')
x = symbols('x')
def bezier(p0, p1, g0, g1):
    a0 = p0[0]
    b0 = p0[1]
    a1 = 3.0*(g0[0] - p0[0])
    b1 = 3.0*(g0[1] - p0[1])
    a2 = 3.0*(p0[0] + g1[0] - 2.0*g0[0])
    b2 = 3.0*(p0[1] + g1[1] - 2.0*g0[1])
    a3 = p1[0] - p0[0] + 3.0*g0[0] - 3.0*g1[0]
    b3 = p1[1] - p0[1] + 3.0*g0[1] - 3.0*g1[1]
    curves = (a0 + a1*t + a2*t**2 + a3*t**3, b0 + b1*t + b2*t**2 + b3*t**3)
    
    
    p = plot_parametric(curves[0], curves[1], (t, 0, 1), show=False)
    p.extend(plot((g0[1] - p0[1]) / (g0[0] - p0[0]) * (x-p0[0]) + p0[1], (x, p0[0], g0[0]), show=False))
    p.extend(plot((g1[1] - p1[1]) / (g1[0] - p1[0]) * (x-p1[0]) + p1[1], (x, p1[0], g1[0]), show=False))
    return curves, p
```

```python
# Problem 1
# a)

curve, p = bezier((0, 0), (5, 2), (1, 1), (6, 1))
print("cubic Hermite approximation:", curve)
p.show()
```

```python
# d)

curve, p = bezier((0, 0), (5, 2), (2, 2), (7, 0))
print("cubic Hermite approximation:", curve)
p.show()
```

```python
# Problem 3
# c)

curve0, p0 = bezier((0, 0), (4, 6), (0.5, 0.5), (3.5, 7))
curve1, p1 = bezier((4, 6), (6, 1), (4.5, 5), (7, 2))

print("cubic Hermite approximation between x0 = 0 and x1 = 4:", curve0)
print("cubic Hermite approximation between x1 = 4 and x2 = 6:", curve1)
p0.extend(p1)
p0.show()
```

```python
# d)

curve0, p0 = bezier((0, 0), (2, 1), (0.5, 0.5), (3, 1))
curve1, p1 = bezier((2, 1), (4, 0), (3, 1), (5, 1))
curve2, p2 = bezier((4, 0), (6, -1), (3, -1), (6.5, -0.25))

print("cubic Hermite approximation between x0 = 0 and x1 = 2:", curve0)
print("cubic Hermite approximation between x1 = 2 and x2 = 4:", curve1)
print("cubic Hermite approximation between x2 = 4 and x3 = 6:", curve2)

p0.extend(p1)
p0.extend(p2)
p0.show()
```
